```python
def show_KDA(data):
    f, axes = plt.subplots(nrows=1, ncols=3, figsize=(14, 6))
    sns.lineplot(x="blueKills", y="blueWins", data=data, ci=95, ax=axes[0])
    sns.lineplot(x="blueDeaths", y="blueWins", data=data, ci=95, ax=axes[1])
    sns.lineplot(x="blueAssists", y="blueWins", data=data, ci=95, ax=axes[2])
    plt.show()
show_KDA(data_clean)
```

```python
def detect_outliers(data, threshold=3):
    mean_d = np.mean(data)
    std_d = np.std(data)
    outliers = []
    r = np.where(np.abs((data - mean_d)/std_d ) > threshold)
    for y in data:
        z_score= (y - mean_d)/std_d 
        if np.abs(z_score) > threshold:
            outliers.append(y)
    return outliers
```

```python
def remove_outliers(data, feature, outliers):
    new_data = data.copy()
    for outlier in outliers:
        new_data = new_data[new_data[feature] != outlier]
    return new_data
```

```python
for feature in ['blueKills', 'blueDeaths', 'blueAssists']:
    outliers = detect_outliers(data_clean[feature])
    data_clean = remove_outliers(data_clean, feature, outliers)
```

```python
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
length = 6000
train = data_clean[:length]
test = data_clean[length:]
X_train = train.drop(["blueWins"], axis=1)
Y_train = train.blueWins
x_train,x_test,y_train,y_test = train_test_split(X_train, Y_train, 
                                                 test_size=0.3, random_state=42)
print("x_train:",len(x_train))
print("x_test:",len(x_test))
print("y_train:",len(y_train))
print("y_test:",len(y_test))
print("test:",len(test))
```

```python
X_train.head()
```

```python
# Logistic Regression
lr = LogisticRegression()
lr.fit(x_train,y_train)
acc_log_train = round(lr.score(x_train,y_train)*100,2)
acc_log_test = round(lr.score(x_test,y_test)*100,2)
print("Training Accuracy: %{}".format(acc_log_train))
print("Test Accuracy: %{}".format(acc_log_test))
```

```python
best_estimators = []
cv_results = []
```

```python
random_state = 42
classifier = LogisticRegression(random_state = random_state)
logreg_param_grid = {"C":np.logspace(-3,3,7),
                    "penalty": ["l1","l2"]}
classifier_param = logreg_param_grid
clf = GridSearchCV(classifier, param_grid=classifier_param, 
                   cv=StratifiedKFold(n_splits = 10), 
                   scoring="accuracy", n_jobs = -1,verbose = 1)
clf.fit(x_train,y_train)
cv_result = clf.best_score_
best_estimators.append(clf.best_estimator_)
print("cross validation result: ", cv_result)
```

```python
lr_acc = accuracy_score(clf.predict(x_test),y_test)
cv_results.append(lr_acc)
print(lr_acc)
```

```python
# Decision Tree  overfit
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
acc_dt_train = round(dt.score(x_train,y_train)*100,2)
acc_dt_test = round(dt.score(x_test,y_test)*100,2)
print("Training Accuracy: %{}".format(acc_dt_train))
print("Test Accuracy: %{}".format(acc_dt_test))
```

```python
random_state = 42
classifier = DecisionTreeClassifier(random_state = random_state)
dt_param_grid = {"min_samples_split" : range(10,500,20),
                "max_depth": range(1,20,2)}
classifier_param = dt_param_grid
clf = GridSearchCV(classifier, param_grid=classifier_param, 
                   cv=StratifiedKFold(n_splits = 10), 
                   scoring="accuracy", n_jobs = -1,verbose = 1)
clf.fit(x_train,y_train)
cv_result = clf.best_score_
best_estimators.append(clf.best_estimator_)
print("cross validation result: ", cv_result)
```

```python
# dt
dt_acc = accuracy_score(clf.predict(x_test),y_test)
cv_results.append(dt_acc)
print(dt_acc)
```

```python
# KNN
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
acc_knn_train = round(knn.score(x_train,y_train)*100,2)
acc_knn_test = round(knn.score(x_test,y_test)*100,2)
print("Training Accuracy: %{}".format(acc_knn_train))
print("Test Accuracy: %{}".format(acc_knn_test))
```

```python
random_state = 42
classifier =  KNeighborsClassifier()
knn_param_grid = {"n_neighbors": np.linspace(1,19,10, dtype = int).tolist(),
                 "weights": ["uniform","distance"],
                 "metric":["euclidean","manhattan"]}
classifier_param = knn_param_grid
clf = GridSearchCV(classifier, param_grid=classifier_param, 
                   cv=StratifiedKFold(n_splits = 10), 
                   scoring="accuracy", n_jobs = -1,verbose = 1)
clf.fit(x_train,y_train)
cv_result = clf.best_score_
best_estimators.append(clf.best_estimator_)
print("cross validation result: ", cv_result)
```

```python
# KNN
knn_acc = accuracy_score(clf.predict(x_test),y_test)
cv_results.append(knn_acc)
print(knn_acc)
```

```python
# ensemble
vc = VotingClassifier(estimators=[("lr",best_estimators[0]),
                                    ("dt",best_estimators[1]),
                                    ("knn",best_estimators[2])],
                                    voting="soft",n_jobs=-1)
vc = vc.fit(x_train,y_train)
vc_acc = accuracy_score(vc.predict(x_test),y_test)
cv_results.append(vc_acc)
print(vc_acc)
```

```python
ensemble = pd.DataFrame({"Cross Validation Means": cv_results,
                         "ML Models":["LogisticRegression",
                                      "DecisionTreeClassifier",
                                      "KNeighborsClassifier",
                                     "EnsembleClassifier"]})

g = sns.barplot("Cross Validation Means","ML Models", data = ensemble)
g.set_xlabel("Mean Acc.")
g.set_title("Cross Validation Scores")
plt.show()
```
